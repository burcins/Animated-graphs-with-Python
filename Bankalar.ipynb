I read all other attachments in a for loop below, however only 1998 data file differentiates from others, so I read it separately and append to others. 

```python
import glob
all_files = glob.glob(os.path.join("*.xls"))
```

```python
all_files.remove("b1998.xls")
```

```python
b = {}
for file in all_files:
    # Getting the file name without extension
    file_name = os.path.basename(file)
    df = pd.read_excel(file_name, sheet_name="Türkçe", skiprows=3)
    df.drop("Unnamed: 0",  axis=1, inplace=True)
    df.dropna(axis=0, inplace=True)
    df.rename(columns={df.columns[0]:'Banka'}, inplace=True)
    df.set_index("Banka", drop=True, inplace=True)
    b[os.path.splitext(os.path.basename(file))[0]] = pd.DataFrame(df)
```

Some of banks were acquired and some of them changed their company titles because of shareholder changes. I changed all of their names related to most current titles, however I may miss some of them. The following codes served this situation. 

```python
for name, df in b.items():
    if any(df.index=='Türkiye Cumhuriyeti Ziraat Bankası'):
        df.rename(index={'Türkiye Cumhuriyeti Ziraat Bankası':'Türkiye Cumhuriyeti Ziraat Bankası A.Ş.'}, inplace=True)
```

```python
for name, df in b.items():
    if any(df.index=='Finans Bank A.Ş.'):
        df.rename(index={'Finans Bank A.Ş.':'QNB Finansbank A.Ş.'}, inplace=True)
```

```python
for name, df in b.items():
    if any(df.index=='Oyak Bank A.Ş.'):
        df.rename(index={'Oyak Bank A.Ş.':'ING Bank A.Ş.'}, inplace=True)
```

```python
for name, df in b.items():
    if any(df.index=='Tekstil Bankası A.Ş.'):
        df.rename(index={'Tekstil Bankası A.Ş.':'ICBC Turkey Bank A.Ş.'}, inplace=True)
```

```python
for name, df in b.items():
    if any(df.index=='Eurobank Tekfen A.Ş.'):
        df.rename(index={'Eurobank Tekfen A.Ş.':'Burgan Bank A.Ş.'}, inplace=True)
```
